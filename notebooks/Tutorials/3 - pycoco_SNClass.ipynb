# The pycoco SNClass
___

This is the workhorse class for manipulating the data in and out of CoCo

```python
%matplotlib notebook
from matplotlib import pyplot as plt

import os
import numpy as np
from astropy.table import Table

import pycoco as pcc
```

```python
sn = pcc.classes.SNClass('SN2007uy')
sn.name
```

## Photometry

```python
sn.load_phot( path = os.path.join(pcc.defaults._default_data_dir_path,"lc/SN2007uy.dat"), verbose = True)
```

```python
sn.phot.data
```

```python
sn.phot.plot()
```

```python
sn.plot_lc()
```

```python
sn.phot.save(filename="SN2007uy_test.dat", path=os.curdir)
```

## Spectra

```python
sn.load_list(os.path.join(pcc.defaults._default_list_dir_path, sn.name+".list"))
sn.load_spec()
```

```python
sn.plot_spec()
```
