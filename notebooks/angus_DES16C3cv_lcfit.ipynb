```python
data_path = "/Users/berto/data/CoreCollapse/crangus_DES/DES16C3cv_orig.dat"
p = pcc.PhotometryClass()
p.load(data_path)
p.plot()
```

```python
T = p._phot_format_for_save()
```

```python
T = T[T.argsort()]
```

```python
pt = T[np.where(T["MJD"] >= 57600)]
```

```python
pt
```

```python
p_short = pcc.PhotometryClass()
p_short.load_table(pt)
```

```python
p_short.plot()
```

```python
p_short.save(filename="DES16C3cv.dat", path = "/Users/berto/data/CoreCollapse/crangus_DES")
```
